```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '../data/processed/patches/train_tier_1/acc/ca041a.h5'
```

```python
with h5py.File(path, 'r') as f:
    images = f['images']  # (N, 512, 512, 3)
    masks  = f['masks']   # (N, 512, 512)
    
    idx = 20001  # change to browse different patches
    img  = images[idx][:]
    mask = masks[idx][:]
    print(f['images'].shape)
    print(f['masks'].shape)
```

21772 tiles of shape 512 x 512. The images have three bands while the masks only have one

```python
img.shape
```

```python
mask.shape
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(img)
axes[0].set_title(f'Image (patch {idx})')
axes[1].imshow(mask, cmap='gray')
axes[1].set_title('Mask')
plt.tight_layout()
plt.show()
```
