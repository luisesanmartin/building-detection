```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
from rasterio.windows import Window
```

```python
tif_path = '../data/raw/train_tier_1/acc/ca041a/ca041a.tif'
mask_path = '../data/processed/train_tier_1/acc/ca041a_mask.tif'
```

```python
with rasterio.open(tif_path) as src:
    rgb = src.read([1, 2, 3])  # R, G, B bands
with rasterio.open(mask_path) as src:
    mask = src.read(1)
```

```python
print(rgb.shape)
print(mask.shape)
```

```python
# Window in pixels. The zeros are col and row offsets
window = Window(10000, 10000, 2000, 2000)
```

```python
with rasterio.open(tif_path) as src:
    rgb = src.read([1, 2, 3], window=window)  # R, G, B bands
with rasterio.open(mask_path) as src:
    mask = src.read(1, window=window)
```

```python
print(rgb.shape)
print(mask.shape)
```

```python
# Normalize RGB for display
rgb = np.moveaxis(rgb, 0, -1)
#rgb = (rgb / rgb.max() * 255).astype(np.uint8)
```

```python
# Visualizing
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].imshow(rgb)
axes[0].set_title('Raster Image (X)')
axes[0].axis('off')

axes[1].imshow(mask, cmap='gray')
axes[1].set_title('Labels')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('acc-ca041a.png', dpi=150, bbox_inches="tight")
plt.show()
```
